```python
import pandas as pd

# 练习1：DataFrame数据访问（第37集重点）
data = {
    '姓名': ['小明', '小红', '小刚', '小芳'],
    '数学': [85, 92, 78, 95],
    '英语': [88, 90, 82, 96],
    '年龄': [18, 19, 18, 20]
}
df = pd.DataFrame(data)

print("原DataFrame：\n", df)

# loc（标签索引）
print("\n小明的数学成绩:", df.loc[0, '数学'])
print("前两行所有成绩:\n", df.loc[:1, ['数学', '英语']])

# iloc（位置索引）
print("\n第3行英语成绩:", df.iloc[2, 2])

# 布尔索引（第37集重点）
print("\n数学成绩>85分的同学:\n", df[df['数学'] > 85])

# 练习2：常见方法（第38集）
print("\n前2行:\n", df.head(2))
print("排序（数学降序）:\n", df.sort_values('数学', ascending=False))
print("删除年龄列:\n", df.drop(columns=['年龄']))

# 练习3：小案例（第39集风格）
# 计算总分并筛选
df['总分'] = df['数学'] + df['英语']
print("\n总分>170分的同学:\n", df[df['总分'] > 170])
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
np.random.seed(42)
n = 200
X = np.random.rand(n,2)
y = (X[:,0] + X[:,1]>1.0).astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"训练集: {len(X_train)} 条，测试集: {len(X_test)} 条")
y_pred = (X_test[:, 0] + X_test[:, 1] > 0.95).astype(int)
print("准确率 Accuracy:", round(accuracy_score(y_test, y_pred), 3))
print("精确率 Precision:", round(precision_score(y_test, y_pred), 3))
print("召回率 Recall:", round(recall_score(y_test, y_pred), 3))
print("F1分数:", round(f1_score(y_test, y_pred), 3))
print("\n混淆矩阵:\n", confusion_matrix(y_test, y_pred))
plt.figure(figsize=(6,4))
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, cmap='coolwarm')
plt.title('测试集真实标签分布（蓝色=正常，红色=患病）')
plt.show()
```
